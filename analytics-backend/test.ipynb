```python
from underthesea import word_tokenize
```

```python
text_list = [word_tokenize(' '.join(text.split())) for text in list(df['User Message'])]
```

```python
text_list
```

```python
import fasttext
model = fasttext.load_model('data/cc.vi.300.bin')
```

```python
import numpy as np
```

```python
def get_sentence_vectors(text_list):
    vectors = [
        np.average(
            np.array([model.get_word_vector(word) for word in sentence])
        , axis=0)
        for sentence in text_list
        if len(sentence) > 0
    ]
    return vectors
```

```python
sent_vectors = get_sentence_vectors(text_list)
```

```python

from sklearn.cluster import DBSCAN
from sklearn import metrics
# from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
```

```python

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
db = DBSCAN().fit(sent_vectors)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print('Estimated number of clusters: %d' % n_clusters_)
print('Estimated number of noise points: %d' % n_noise_)
```

```python
import matplotlib.pyplot as plt

# Black removed and is used for noise instead.
unique_labels = set(labels)
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = (labels == k)
    X = np.array(sent_vectors)

    xy = X[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=14)

    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=6)

plt.title('Estimated number of clusters: %d' % n_clusters_)
plt.show()
```
