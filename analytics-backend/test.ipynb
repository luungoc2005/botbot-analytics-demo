```python
df['MessageReceived'] = pd.to_datetime(df['MessageReceived'])
df = df.set_index(pd.DatetimeIndex(df['MessageReceived']))
```

```python
df.head()
```

```python
tmp_df = df.copy()
tmp_df['User Message Count'] = tmp_df.apply(lambda x: x['User Message'].count('thu'), axis=1)
```

```python
res = tmp_df.groupby(df.index.to_period('D')).sum()[['User Message Count']]
```

```python
res.index = pd.to_datetime(res.index.to_timestamp())
res.head()
```

```python
res.index = res.apply(lambda x: x.index.strftime("%m/%d/%Y %H:%M:%S"))
```
